```python
#Hacer un programa que ingrese por teclado un número inicial y un número
#final para generar un rango (una la lista consecutiva de números que inicia y
#termina de acuerdo a los números ingresados):

# Función para verificar si un número es par o impar
def verificar_paridad(numero):
    if numero % 2 == 0:
        return "par"
    else:
        return "impar"

# Función para generar el rango y mostrar la paridad de cada número
def procesar_rango(inicio, fin):
    # Generar la lista del rango de números
    numeros = list(range(inicio, fin + 1))
    
    # Recorrer el rango e imprimir si el número es par o impar
    for numero in numeros:
        paridad = verificar_paridad(numero)
        print(f"El número {numero} es {paridad}.")

# Solicitar al usuario el número inicial y final
try:
    inicio = int(input("Ingrese el número inicial: "))
    fin = int(input("Ingrese el número final: "))
    
    if inicio > fin:
        print("El número inicial debe ser menor o igual al número final.")
    else:
        procesar_rango(inicio, fin)
except ValueError:
    print("Por favor, ingrese valores numéricos válidos.")

```

```python
#Hacer un programa que imprima la tabla completa de multiplicar de un
#número que se ingresa por teclado:

# Función para imprimir la tabla de multiplicar de un número
def imprimir_tabla(numero):
    print(f"Tabla de multiplicar del {numero}:")
    for i in range(1, 11):  # Recorre del 1 al 10
        print(f"{numero} x {i} = {numero * i}")

# Solicitar al usuario un número
try:
    numero = int(input("Ingrese un número para imprimir sus tablas de multiplicar del 1 al 10: "))
    imprimir_tabla(numero)
except ValueError:
    print("Por favor, ingrese un número válido.")
```

```python
#Hacer un programa que solicite el ingreso de un nombre y apellido:
# Función para verificar si un texto es válido (sin números ni símbolos extraños)
def es_valido(texto):
    # Listas de caracteres no permitidos
    numeros = "0123456789"
    simbolos = "!@#$%^&*()-_=+[]{}|;:'\",.<>?/\\`~"
    
    # Verificar cada carácter en el texto
    for char in texto:
        if char in numeros or char in simbolos:
            return False  # Retorna False si encuentra un carácter inválido
    return True  # Retorna True si es válido

# Función para solicitar y validar nombre o apellido
def solicitar_texto(mensaje):
    while True:
        texto = input(mensaje).strip()  # Solicita texto y elimina espacios extra
        if es_valido(texto):
            return texto  # Devuelve el texto si es válido
        print("Entrada inválida. No debe contener números ni símbolos extraños. Inténtelo de nuevo.")

# Solicitar y validar nombre y apellido
nombre = solicitar_texto("Ingrese su nombre: ")
apellido = solicitar_texto("Ingrese su apellido: ")

# Mostrar resultado final
print(f"Nombre y apellido válidos: {nombre} {apellido}")
```

```python
#Hacer un programa que verifique los comentarios de los alumnos EPIT:

# Función para verificar el tipo de comentario
def clasificar_comentario(comentario):
    # Listas de palabras negativas y positivas
    palabras_negativas = ["malo", "pésimo", "horrible", "no", "terrible", "fatal"]
    palabras_positivas = ["bueno", "excelente", "bien", "genial", "fantástico", "positivo"]
    
    # Convertir comentario a minúsculas para comparación
    comentario = comentario.lower()
    
    # Verificar si contiene palabras negativas
    for palabra in palabras_negativas:
        if palabra in comentario:
            return "Comentario NEGATIVO"
    
    # Verificar si contiene palabras positivas
    for palabra in palabras_positivas:
        if palabra in comentario:
            return "Comentario POSITIVO"
    
    # Si no encuentra palabras negativas ni positivas
    return "Comentario NEUTRO"

# Solicitar el comentario al alumno
comentario = input("Ingrese su comentario: ")

# Clasificar y mostrar el resultado
resultado = clasificar_comentario(comentario)
print(resultado)
```

```python
#Hacer un programa que genere una lista aleatoria de hasta 60 numeros:

import random

# Función para generar una lista aleatoria
def generar_lista_aleatoria(tamaño, maximo):
    return [random.randint(1, maximo) for _ in range(tamaño)]

# Función para clasificar los números en listas de múltiplos
def clasificar_multiplos(lista):
    multiplos_2 = [num for num in lista if num % 2 == 0]
    multiplos_3 = [num for num in lista if num % 3 == 0]
    multiplos_5 = [num for num in lista if num % 5 == 0]
    return multiplos_2, multiplos_3, multiplos_5

# Generar la lista aleatoria
tamaño_lista = random.randint(10, 60)  # Tamaño entre 10 y 60
maximo_valor = 100  # Los números estarán entre 1 y 100
lista_numeros = generar_lista_aleatoria(tamaño_lista, maximo_valor)

# Clasificar los números en múltiplos de 2, 3 y 5
multiplos_2, multiplos_3, multiplos_5 = clasificar_multiplos(lista_numeros)

# Mostrar resultados
print("Lista original:")
print(lista_numeros)
print("\nMúltiplos de 2:")
print(multiplos_2)
print("\nMúltiplos de 3:")
print(multiplos_3)
print("\nMúltiplos de 5:")
print(multiplos_5)

```

```python
#Hacer un programa que ingrese por teclado un número inicial de 3 cifras y
#un número final de 4 cifras para generar un rango (una la lista consecutiva de
#números que inicia y termina de acuerdo a los números ingresados):

# Función para verificar si un número es capicúa
def es_capicua(numero):
    numero_str = str(numero)  # Convertir el número a cadena
    return numero_str == numero_str[::-1]  # Comparar con su inverso

# Función para procesar un rango y verificar si los números son capicúa
def verificar_capicuas(inicio, fin):
    lista_numeros = list(range(inicio, fin + 1))  # Generar el rango
    print(f"Rango generado: {lista_numeros}")
    
    # Recorrer la lista y verificar cada número
    for numero in lista_numeros:
        if es_capicua(numero):
            print(f"{numero} es Capicúa")
        else:
            print(f"{numero} no es Capicúa")

# Solicitar datos al usuario
try:
    inicio = int(input("Ingrese el número inicial (de 3 cifras): "))
    fin = int(input("Ingrese el número final (de 4 cifras): "))
    
    # Validar las entradas según el requerimiento
    if len(str(inicio)) == 3 and len(str(fin)) == 4 and inicio <= fin:
        verificar_capicuas(inicio, fin)
    else:
        print("Error: El número inicial debe tener 3 cifras, el final debe tener 4 cifras, y el rango debe ser válido.")
except ValueError:
    print("Por favor, ingrese valores numéricos válidos.")
```

```python
#Hacer un programa que solicite el ingreso del DNI de un ciudadano:
#Que verifique que el DNI no contenga letras del abecedario.
#Que verifique que el DNI no contenga ningun simbolo extraño.
#Que verifique que el DNI solo tenga 8 números.
#Si detecta que el DNI no es válido, que solicite que vuelva a ingresar su DNI.
#Utilizar for y funciones 

# Función para verificar que el DNI no contenga letras
def verificar_sin_letras(dni):
    for caracter in dni:
        if caracter.isalpha():  # Verifica si el caracter es una letra
            return False
    return True

# Función para verificar que el DNI no contenga símbolos extraños
def verificar_sin_simbolos(dni):
    for caracter in dni:
        if not caracter.isdigit():  # Verifica si el caracter no es un número
            return False
    return True

# Función para verificar que el DNI tenga exactamente 8 números
def verificar_8_numeros(dni):
    if len(dni) == 8 and dni.isdigit():  # Verifica que tenga 8 caracteres y todos sean dígitos
        return True
    return False

# Función principal para solicitar el ingreso del DNI
def solicitar_dni():
    while True:
        dni = input("Por favor, ingrese su DNI (8 dígitos): ")
        
        # Verificar que el DNI no contenga letras
        if not verificar_sin_letras(dni):
            print("El DNI no puede contener letras. Intente nuevamente.")
            continue
        
        # Verificar que el DNI no contenga símbolos extraños
        if not verificar_sin_simbolos(dni):
            print("El DNI no puede contener símbolos extraños. Intente nuevamente.")
            continue
        
        # Verificar que el DNI tenga exactamente 8 números
        if not verificar_8_numeros(dni):
            print("El DNI debe tener exactamente 8 números. Intente nuevamente.")
            continue
        
        print("DNI ingresado correctamente.")
        break  # Salir del bucle cuando el DNI es válido

# Llamada a la función principal para ejecutar el programa
solicitar_dni()
```
